```python
URL = "https://www.wetalk.tw/forum.php?mod=forumdisplay&fid=2&page=1"
resp = requests.get(URL)
soup = BS(resp.text, 'html5lib')
str(soup)
```

```python
a_tags = [a.get('href') for a in soup.select('tbody > tr > th > a.common')]
a_tags += [a.get('href') for a in soup.select('tbody > tr > th > a.xst')]
a_tags = a_tags[2:]
a_tags
```

```python
for a_tag in a_tags:
    resp = requests.get(a_tag)
    soup = BS(resp.text, 'html5lib')
    soup
```
