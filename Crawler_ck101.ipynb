```python
import requests
import re
import json
from bs4 import BeautifulSoup as BS
```

```python
# 卡提諾 政治討論 共241頁 可追溯回2016-5-4 09:39
```

```python
for page in range(1, 2): #正是要爬時改成242
    URL = "https://ck101.com/forum-3586-%s.html"%page
    header = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36'}
    resp = requests.get(URL, headers=header)
    soup = BS(resp.text, 'html5lib')
    a_tags = [a.get('href') for a in soup.select('a.s.xst')]    

    for a_tag in a_tags:
        resp = requests.get(a_tag, headers=header)
        soup = BS(resp.text, 'html5lib')
        
        _id = re.sub('https://ck101.com/thread-|-1-1.html', '', a_tag )
#         print(_id)
        title = soup.select('hgroup > h1')[0].text
        content = soup.select('td.t_f')[0].text
        content = re.sub('\n| |Star|以上來自.+', '', content)

        news_dict = {}
        news_dict['id'] = _id
        news_dict['title'] = title
        news_dict['content'] = content
#         print(news_dict)
        
        with open("./News_Crawler/ck101/ck101_%s.json"%_id, 'w') as f:
            jd = json.dumps(news_dict, ensure_ascii=False)
            jd.encode("utf-8")
            f.write(jd)
        
```
